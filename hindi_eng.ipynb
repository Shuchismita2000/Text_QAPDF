```python
with open('63 Psychology.txt', 'r', encoding='utf-8') as f:
    text=[]
    for line in f:
        splits = line.split()
        if splits !=[]:
            text.append(splits)
```

```python
def req_list(req_string):
    index=[]
    for item in text:
        x=item
        if (set(req_string).issubset(set(x))):
            index.append(text.index(item))
    return index    
```

```python
import re
digit=[]
sub=[]
for i in g_i:
    s=text[i]
    for x in s:
        p=re.compile(r'\d+\-\d+')
        q=p.findall(x)
        if q!=[]:
            digit.append(q)
        if x.isdigit()==True:
            sub.append(x)
digit.append(sub)
```
